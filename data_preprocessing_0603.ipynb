```python
def shift_process(df, col, n):
    """
    a function that process a DataFrame, adding previous data given last n year
    df: pandas.DataFrame
    col: a column that need to be processed
    n: n year before
    """
    extra_col = df[col].shift(periods=n)
    df[f'{col}_pass{n}'] = extra_col
    return df
```

```python
for i in range(1, 6):
    df = shift_process(df, 'sale', i)
for i in range(1, 6):
    df = shift_process(df, 'xrd', i)
df.head(3)
```

```python
# testing
test_df = df[df.conm == 'REDHILL BIOPHARMA LTD']
test_row = test_df.iloc[5, :]
test_row
```

### Calculate Ability

```python
# create functions for calculating ability
def calculate_log_sale(data, col, n):
    return [np.log(data[col] / data[f'{col}_pass{i}']) for i in range(1, n+1)]

def calculate_rd(data, col, n):
    return [np.log(1 + data[f'{col}_pass{i}']) for i in range(1, n+1)]


d = {'X': calculate_rd(test_row, 'xrd', 5), 'y': calculate_log_sale(test_row, 'sale', 5)}
df = pd.DataFrame(data=d)
df
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# statsmodels
model = smf.ols(formula='y ~ X', data=df)

# Fit the regression
regression = model.fit()
regression.params
```
